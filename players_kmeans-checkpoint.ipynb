```python
import numpy as np
import pandas as pd 
```

```python
players = pd.read_csv('players_21.csv')
```

```python
players.head()
```

```python
features = ['overall','potential','wage_eur','value_eur','age']
players = players.dropna(subset=features)
```

```python
data = players[features].copy()
```

```python
# Scaling data to standardize values 
# scaling the values from 1 to 10
data = ((data- data.min())/(data.max()-data.min()))*9 + 1 
```

```python
data.describe
```

```python
data.head()
```

```python
# selecting random centroids for K means algo .
def random_centroids(data,k):
    centroids = []
    for i in range(k):
        centroid = data.apply(lambda x : float(x.sample()))
        centroids.append(centroid)
    return pd.concat(centroids,axis=1)
```

```python
centroids = random_centroids(data,5)
centroids
```

```python
def get_labels(centroids,data):
    distances = centroids.apply(lambda x : np.sqrt(((data - x)**2).sum(axis=1)))
    # distances
    return distances.idxmin(axis=1)
```

```python
labels = get_labels(centroids,data)
```

```python
labels
```

```python
def new_centroids(data,labels,k):
    return data.groupby(labels,group_keys=True).apply(lambda x : np.exp(np.log(x).mean())).T
```

```python
from sklearn.decomposition import PCA # Principle component analysis for visualisastion
import matplotlib.pyplot as plt 
from IPython.display import clear_output
```

```python
def plot_clusters(data, labels, centroids, iteration):
    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)
    centroids_2d = pca.transform(centroids.T)
    clear_output(wait=True)
    plt.title(f'Iteration {iteration}')
#     print(labels)
    plt.scatter(x=data_2d[:,0], y=data_2d[:,1], c=labels.astype(float))
    plt.scatter(x=centroids_2d[:,0], y=centroids_2d[:,1],c='red')
    plt.show()
```

```python
max_iterations = 100
k = 3
centroids = random_centroids(data,k)

old_centroids = pd.DataFrame()
iteration = 1 
while iteration < max_iterations and not centroids.equals(old_centroids):
    old_centroids = centroids
    distances = centroids.apply(lambda x : np.sqrt(((data - x)**2).sum(axis=1)))
    labels= distances.idxmin(axis=1)
#     labels=get_labels(data,centroids)
    print(labels)
    centroids = new_centroids(data,labels,k)
#     print(centroids_n)
    plot_clusters(data,labels,centroids,iteration)
    iteration+= 1 
```
